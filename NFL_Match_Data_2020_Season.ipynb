# NFL Match Data - 2020 Season Analysis

```python
import pandas as pd
```

```python
df = pd.read_csv("Data/processed_game_history.csv")
df = df.drop(["Unnamed: 0", "AtHome"], axis=1)
```

```python
df_2020 = df[df["Season"] == 2020]
print(f"Number of games played: {len(df_2020)}")
```

```python
df_2020.head()
```

```python
wk_occurence_2020 = df_2020.groupby(["Week"]).size().reset_index().rename(columns={0:"count"}).sort_values(by="count", ascending=False)
wk_occurence_2020
```
